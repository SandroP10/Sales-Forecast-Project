```python
import matplotlib.pyplot as plt

# Convert date columns to datetime
df['yearweek_start'] = pd.to_datetime(df['yearweek_start'])
df['yearweek_end'] = pd.to_datetime(df['yearweek_end'])

# Set index to start date
df = df.set_index('yearweek_start')

# Resample to monthly frequency
monthly_sales = df.resample('M')['sales'].sum()

# Plot the timeseries
fig, ax = plt.subplots()
monthly_sales.plot(ax=ax)

ax.set_xlabel('Date')
ax.set_ylabel('Monthly Sales')
ax.set_title('Monthly Sales Time Series')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.stats import zscore

z = zscore(df['sales'])
df[np.abs(z) > 3] # potential outliers
```

The dates in late 2020 such as 11/30/2020, 11/2/2020, 10/26/2020 have very high positive z-scores, indicating they are potential outliers on the higher end of the sales distribution. This aligns with the times of year when sales are usually elevated.

Dates like 3/9/2020 and 4/27/2020 also have high z-scores, suggesting anomalously high sales for those periods that are outliers compared to typical levels.

On the low end, dates like 1/17/2022, 1/31/2022, 3/20/2023 have very negative z-scores, flagging them as potential low outliers.

The low z-score dates seem to be occasional throughout the year, while high z-scores cluster at end of year holidays.

```python
# Filter for outlier rows
df['zscore'] = zscore(df['sales'])
```

```python
outliers = df[np.abs(df['zscore']) > 3]
```

```python
Q1 = df['sales'].quantile(0.25)
Q3 = df['sales'].quantile(0.75)
IQR = Q3 - Q1

df[~((df['sales'] >= Q1 - 1.5 * IQR) & (df['sales'] <= Q3 + 1.5 * IQR))]
```

The dates flagged as high outliers seem concentrated at the end of year holiday periods - November/December primarily.
sales are expected to be higher.

Specific dates like 11/30/2020 and 12/7/2020 have sales above Q3 + 1.5*IQR, indicating anomalously high values.
No dates are being flagged as low outliers below Q1 - 1.5*IQR

The presence of outliers on the high end makes sense for retail sales data, where seasonal events can disproportionately spike sales.

```python
# Sine/cosine encoding (Cyclical Encoding)
df['month_sin'] = np.sin(2 * np.pi * df['month']/12)
df['month_cos'] = np.cos(2 * np.pi * df['month']/12)
```

```python
# Extract day of week
df['day_of_week'] = df.index.dayofweek
df['day_of_week_sin'] = np.sin(2 * np.pi * df['day_of_week']/7)
df['day_of_week_cos'] = np.cos(2 * np.pi * df['day_of_week']/7)
```

This encodes monthly and weekly seasonality into sinusoidal features.

```python
# Rolling mean on lag-1 sales
df['lag1_sales_7d_mean'] = df['sales_lag1'].rolling(7).mean()

# Rolling std on lag-2 sales
df['lag2_sales_30d_std'] = df['sales_lag2'].rolling(30).std()
```

Rolling windows help capture trends and variation over recent periods.

```python
df['days_since_promo'] = df['promo_01'].replace(0,np.nan).fillna(method='ffill').fillna(0).astype(int)
```

This counts days since the last promotion.

```python
df['month_promo01'] = df['month'] * df['promo_01']
df['day_promo02'] = df['day_of_week'] * df['promo_02']

#Interactions between promos and month/day
```
